```python
# fonction generique pour histogram
def Histogram(data,to_plot):
    for i in range(len(to_plot)):
        plt.hist(data[to_plot[i]])
        plt.axvline(data[to_plot[i]].mean(),color='r')
        plt.xlabel(to_plot[i])
        plt.show()
```

```python
to_plt =['Pclass','Fare']
Histogram(train_df,to_plt)
```
